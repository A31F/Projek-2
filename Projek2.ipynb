```python
#membaca folder csv
suhu = pd.read_csv("suhu 1.csv")
#memberi nomer urut
suhu = suhu.set_index(pd.Index(range(1, len(suhu)+1)))
#menampilkan
suhu
```

```python
tipe = suhu['WAKTU']
asimilasi = suhu['Asimilasi']
observasi = suhu['Observasi']
non_asimilasi = suhu['Non Asimilasi']

# Plotting
plt.figure(figsize=(13, 5))
plt.plot(tipe, asimilasi, color='blue', label='Asimilasi',marker='*',zorder=2)
plt.plot(tipe, observasi, color='yellow', label='Observasi',marker='*',zorder=2)
plt.plot(tipe, non_asimilasi, color='red', label='Non Asimilasi',marker='*',zorder=2)

for temp in range(20, 28):
    plt.axhline(y=temp, color='gray', linestyle='-', linewidth=0.5,zorder=1)

# Kustomisasi
plt.xlabel('Waktu')
plt.ylabel('Suhu (°C)')
plt.title('Grafik Suhu')
plt.xticks(rotation=15)
plt.ylim(20, 27)

plt.legend()


# Tampilkan plot
plt.show()












# Afief Dwi Hidayat (0075)
```
